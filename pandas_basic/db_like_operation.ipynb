```python
import pandas as pd
```

```python
dset1 = pd.read_excel('data_set1.xlsx')
dset1
```

```python
dset2 = pd.read_excel('data_set2.xlsx')
dset2
```

#### 1. 根据两个字段merge;
#### 2. 使用`indicator`参数显示每一行记录的来源;
#### 3. 使用`suffixes`参数为两个表中的重复字段添加后缀
- https://stackoverflow.com/a/47107164/2803344
- https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html

```python
dset_merged = pd.merge(dset1, dset2, left_on=['cid', 'adducts'], right_on=['cid', 'adducts'], 
                       how='left', indicator=True, suffixes=['', '_R'])
dset_merged
```

#### 4. 使用`query`对DF进行过滤
- 下面可以得到data set1中有但是data set2中没有的数据
- https://stackoverflow.com/a/44546263/2803344

```python
dset_merged.query("_merge=='left_only'")
```
