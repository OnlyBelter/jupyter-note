### Query LIPID MAPS
- http://www.lipidmaps.org/data/lm_rest.php

```python
import os
import requests
import pandas as pd
import numpy as np
```

```python
def query_by_name(base_url, context, q_from, q_value, q_to, output_format):
    query = os.path.join(base_url, context, q_from, q_value, q_to, output_format)
    r = requests.get(query)
    if r.status_code == 200:
        return r.content
    else:
        return None
```

```python
BASE_URL = r'http://www.lipidmaps.org/rest'
context = 'compound'
q_from = 'abbrev'
q_to = 'pubchem_cid'
output_format = 'txt'
```

```python
q_value = "(3'-sulfo)Galbeta-Cer(d18:1/24:0)"
```

```python
q_result = query_by_name(base_url=BASE_URL, context=context, q_from=q_from, q_value=q_value, q_to=q_to, output_format=output_format)
```
