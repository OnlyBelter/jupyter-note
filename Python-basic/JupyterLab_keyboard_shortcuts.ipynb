```python
import requests
import pandas as pd

url = "https://raw.githubusercontent.com/jupyterlab/jupyterlab/master/packages/shortcuts-extension/schema/plugin.json"
resp = requests.get(url=url)
data = resp.json()

df = pd.DataFrame(data["properties"])
df = pd.concat(df.loc["properties"].apply(lambda x: pd.DataFrame(x)).values)
df = df.set_index(["category", "title"]).sort_index()[["keys"]]

with pd.option_context("display.max_rows",100):
    display(df)
```
